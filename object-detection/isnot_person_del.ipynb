```python
import os

path = "car/images/"
label_list_path = os.listdir(path)
label_list = [file for file in label_list_path if file.endswith(".txt")]
print(label_list)
print("\n\n")

for file in label_list:
    person = False 
    f = open(path+file,'r')
    while True:
        line = f.readline()
        if not line: 
            break
        #     
        if line[0] == '0':
            person = True
            print(file)
            break
    f.close()
    #    
    if person == False:
         os.remove(path+file) 
    
```
