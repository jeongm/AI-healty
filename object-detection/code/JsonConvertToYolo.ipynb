```python
# 폴더 내 모든 json파일 목록 가져오기
import glob
import os, shutil
import json

food_path = '../건강관리를 위한 음식 이미지/Training_images/'
label_path = '../건강관리를 위한 음식 이미지/Training_labels/'
move_path = '../건강관리를 위한 음식 이미지/json/'
# 라벨 
file_list = os.listdir(label_path) # 모든 파일 다 가져옴
json_list = [file for file in file_list if file.endswith('.json')] # json 파일만 가져옴
print(len(json_list))

# 이미지
food_list = os.listdir(food_path) # 이미지 폴더 
print(len(food_list))


# 1. 이미지 폴더 내로 들어감
#    이미지 리스트 가져와
# 2. 라벨 폴더에서 


save_path = '../건강관리를 위한 음식 이미지/labels/'
label = 0 # 라벨 값 주자
img_cnt =0 # 이미지 개수 확인
for food in food_list:
    img_list = os.listdir(food_path + food)
    check = 0
    img_cnt += len(img_list)
    
    print(food + ' start')
    for file in json_list:
        file_check = file[:-5] + '.jpg'
        if file_check in img_list:
            with open(label_path + file,'r') as f:
                json_data = json.load(f) 

            # 좌표값 추출
            point = json_data[0]['Point(x,y)'].split(',')
            x = point[0]
            y = point[1]

            label_in=[]
            label_s = str(label)
            label_in.append(label_s) # 라벨값 
            label_in.append(x)
            label_in.append(y)
            label_in.append(json_data[0]['W'])
            label_in.append(json_data[0]['H'])
            #print(file ,label_in)

            file_name = file[:-5]
            # list txt파일로 저장
            with open(save_path+file_name+'.txt','w',encoding='UTF-8') as f:
                for value in label_in:
                    f.write(value+' ')

            check +=1
            shutil.move(label_path + file, move_path)
    label+=1
    print(food + '  ' + str(label) + '  '+ str(check) + '  fin')
```

# 음식 영양정보- 한식 -라벨링

```python
# 폴더 내 모든 txt파일 목록 가져오기 -> 0 삭제
import glob
import os, shutil

label_path = 'C:\\Users\\user\yolov5\\data\\test_datasets2\\val\\labels'

# 라벨 
label_list = os.listdir(label_path) # 모든 파일 다 가져옴
print(len(label_list))

label = -1 # 라벨 값 주자
point = ''
print(point)
for file in label_list:
    if file[:15] != point:
        point = file[:15]
        label +=1
        print(label)
        print(point +' start')
    with open(label_path +'/'+ file,'r') as f:
        value = f.read().split('\n')
        #print(value)
        for v in value:
            if v == '':
                pass
            elif v[0] != '1':
                value.remove(v)
        #print(value)
    with open(label_path+'/'+file,'w',encoding='UTF-8') as f:
        for v in value:
            f.write(v+' ')
print('-fin-')    
```

```python
# 폴더 내 모든 json파일 목록 가져오기
import glob
import os, shutil

label_path = 'C:\\Users\\user\yolov5\\data\\health_datasets\\train\\labels\\'

# 라벨 
label_list = os.listdir(label_path) # 모든 파일 다 가져옴
print(len(label_list))

label = -1 # 라벨 값 주자
pointer = 'start'
cnt = 0
for file in label_list:
    if file[:7] != pointer:
        print(pointer + '  ' + str(cnt) + '  ' + str(label) + '  fin')
        pointer=file[:7]
        print(pointer + '  start')
        cnt=0
        label+=1
    with open(label_path +'/'+ file,'r') as f:
        value = f.read().split(' ')
        #print(value)
        del value[0]
        value.insert(0, str(label))
        #print(value)
    cnt+=1
    with open(label_path+file,'w',encoding='UTF-8') as f:
        for v in value:
            f.write(v+' ')
print('\n-fin-')    
```
