```python
import pandas as pd
```

```python
# 데이터 불러오기
file_path = "../Crawling_Data.txt"

with open(file_path,'r',encoding='UTF-8') as f:
    craw_list = f.read().splitlines()

print(craw_list,len(craw_list))
```

```python
file_path = "../Standard_CodeName.txt"

with open(file_path,'r',encoding='UTF-8') as f:
    img_list = f.read().splitlines()

print(img_list,len(img_list))
```

```python
# 이미지 데이터와 크롤링 데이터 중 겹치는 메뉴만
same_menu=[]

for menu in craw_list:
    if menu in img_list:
        same_menu.append(menu)
print(same_menu, len(same_menu))

# 크롤링 데이터 중 겹치는 메뉴 제거(python list 중복 제거)
sm_set = set(same_menu)
same_menu = list(sm_set)
print(same_menu, len(same_menu))
```

```python
# txt파일로 저장
same_menu.sort()

filePath = '../same_data_origin.txt'
with open(filePath,'w',encoding='UTF-8') as f:
    for menu in same_menu:
        f.write(menu+'\n')
```
