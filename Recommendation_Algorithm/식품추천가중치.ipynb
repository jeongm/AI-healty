```python
import pandas as pd

df = pd.read_excel("식단예시2차.xlsx",header=0,index_col="Unnamed: 0")
df.head()
```

df['가중치'] = 1.0

```python
df.head()
```

```python
W = pd.DataFrame(df['식품상세분류'].unique())
W.head()
```

```python
W['가중치']= 1.0
W.head()
```

```python
s_list = W.sample(n=5,weights="가중치")[0].values
s_list
```

```python
print(W[0])
```

```python
# 가중치 업데이트 코드 1회
while(len(W[W['가중치']>5])<=2):
    select = pd.DataFrame()
    for li in s_list:
        select = select.append(df[df['식품상세분류']==li].sample(n=1))

    print(select['식품명'])
    answer = int(input("음식번호를 입력해주세요 : "))
    W.loc[W[W[0] == df.iloc[answer]['식품상세분류']]['가중치'].index, '가중치'] = W[W[0]==df.iloc[answer]['식품상세분류']]['가중치'].add(W[W[0]==df.iloc[answer]['식품상세분류']]['가중치']/2+1)
```
