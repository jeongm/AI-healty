```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

browser = webdriver.Chrome("c:/temp/chromedriver_win32/chromedriver.exe")
browser.get("https://map.naver.com/v5/")
browser.implicitly_wait(10)
browser.maximize_window()

search = browser.find_element_by_css_selector("input.input_search")
search.click()
time.sleep(1)
search.send_keys("음식점")
time.sleep(1)
search.send_keys(Keys.ENTER)
time.sleep(2)

browser.switch_to.frame("searchIframe")

browser.find_element_by_css_selector("#_pcmap_list_scroll_container").click()

name_data = []
star_data = []
visit_data = []
blog_data = []

for i in range(5):
    lis = browser.find_elements_by_css_selector("li._1EKsQ")
    before_len = len(lis)
    while True:

        browser.find_element_by_css_selector("body").send_keys(Keys.END)

        time.sleep(1.5)

        lis = browser.find_elements_by_css_selector("li._1EKsQ")
        after_len = len(lis)


        if before_len == after_len:
            break
        before_len = after_len


    browser.implicitly_wait(0)


    for li in lis:
        if len(li.find_elements_by_css_selector("span._2FqTn._1mRAM > em")) > 0:
            # 가게명
            name = li.find_element_by_css_selector("span.OXiLu").text
            # 별점
            star = li.find_element_by_css_selector("span._2FqTn._1mRAM > em").text

            if len(li.find_elements_by_css_selector("span._2FqTn._4DbfT")) > 0:
                try:
                    visit_review = li.find_element_by_css_selector("span._2FqTn:nth-child(3)").text
                except:
                    visit_review = "방문자리뷰 0"
                try:
                    blog_review = li.find_element_by_css_selector("span._2FqTn:nth-child(4)").text
                except:
                    blog_review = "블로그리뷰 0"
            else:
                try:
                    visit_review = li.find_element_by_css_selector("span._2FqTn:nth-child(2)").text
                except:
                    visit_review = "방문자리뷰 0"
                try:
                    blog_review = li.find_element_by_css_selector("span._2FqTn:nth-child(3)").text
                except:
                    blog_review = "블로그리뷰 0"
            name_data.append(name)
            star_data.append(star)
            visit_data.append(visit_review)
            blog_data.append(blog_review)
            print(name, star, visit_review, blog_review)
    nextp = browser.find_element_by_xpath("//*[@id=\"app-root\"]/div/div[2]/div[2]/a[7]")
    nextp.click()
```

```python
browser = webdriver.Chrome("c:/temp/chromedriver_win32/chromedriver.exe")
browser.get("https://map.naver.com/v5/")
browser.implicitly_wait(10)
browser.maximize_window()

dic_rest = {}

for rest_name in name_data[170:]:
    # 가게이름 검색
    search = browser.find_element_by_css_selector("input.input_search")
    search.click()
    time.sleep(1)
    search.clear()
    search.send_keys(rest_name)
    time.sleep(1)
    search.send_keys(Keys.ENTER)
    time.sleep(2)
    
    # 메뉴창 클릭
    try:
        browser.switch_to.frame("entryIframe")
        time.sleep(1)
        try:
            browser.find_element_by_link_text("메뉴").click()
        except:
            browser.switch_to_default_content()
            continue    
    except:
        browser.switch_to.frame("searchIframe")
        time.sleep(1)
        place = browser.find_element_by_class_name("place_didmount")
        try:
            place.find_element_by_class_name("_2w9xx").click()
        except:
            browser.switch_to_default_content()
            continue
            #place.find_element_by_xpath("//*[@id=\"_pcmap_list_scroll_container\"]/ul/li[3]/div[1]/a/div[1]/div/span[1]").click()
        browser.switch_to_default_content()
        browser.switch_to.frame("entryIframe")
        browser.find_element_by_link_text("메뉴").click()
    time.sleep(1)
    
    # 사진 없는 음식점 예외처리
    try:
        menu = browser.find_element_by_class_name("place_section_content")
        li = menu.find_elements_by_tag_name("li")
    except:
        menu = browser.find_element_by_tag_name("body")
        li = menu.find_elements_by_tag_name("li")
    
    menu_list = []
    for i in li:
        menu_list.append(i.text)
    
    dic_rest[rest_name] = menu_list
    
    browser.switch_to_default_content()
```
