```python
def MouseLeftClick(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        clicked_points.append((y, x))

        image = clone.copy()
        for point in clicked_points:
            cv2.circle(image, (point[1], point[0]), 2, (0, 255, 255), thickness = -1)
        cv2.imshow("image", image)
```

```python
dir_del = None
clicked_points = []
clone = None

path = 'image'
image_ = '220414_141727.png'

now = datetime.now()
now_str = "%s%02d%02d_%02d%02d%02d" % (now.year - 2000, now.month, now.day, now.hour, now.minute, now.second) 

cv2.namedWindow("image")
cv2.setMouseCallback("image", MouseLeftClick)

image = cv2.imread(path+'/'+image_)

clone = image.copy()

flag = False

while True:
    cv2.imshow("image", image)
    key = cv2.waitKey(0)

    if key == ord('n'):
        file_write = open(path+'/' + now_str + '_' + image_[:-5] + '.txt', 'a+')

        text_output = 'point : '+str(len(clicked_points))+'\n'
        for points in clicked_points:
            text_output += str(points[0]) + "," + str(points[1]) + '\n'
        text_output += '\n'
        file_write.write(text_output)

        clicked_points = []

        file_write.close()

        break

    if key == ord('b'):
        if len(clicked_points) > 0:
            clicked_points.pop()
            image = clone.copy()
            for point in clicked_points:
                cv2.circle(image, (point[1], point[0]), 2, (0, 255, 255), thickness = -1)
            cv2.imshow("image", image)

    if key == ord('q'):
        flag = True
        break

    if flag:
        break

cv2.destroyAllWindows()
```

```python
import math
```

```python
x1,y1 = clicked_points[0]
x2,y2 = clicked_points[1]
```

```python
a = x1 - x2    # 선 a의 길이
b = y1 - y2    # 선 b의 길이
 
c = math.sqrt((a * a) + (b * b))  
```

```python
c
```

### 위 두 점 p1, p2는 각각 고운관의 위쪽 꼭짓점으로 측정거리는 네이버 지도 기준 98미터 구글맵 기준 96미터 카카오맵 기준 98미터로 98미터로 두고 계산하였다.

```python
pixelD = 98/c
```
